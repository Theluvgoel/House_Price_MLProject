```python
labels = data['price']
train1 = data.drop(["date","price","street","city","statezip","country"],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test = train_test_split(train1 , labels , test_size = 0.2,random_state =2)
x_train
```

```python
regression_model = LinearRegression()
regression_model.fit(x_train, y_train)
```

```python
for idx, col_name in enumerate(x_train.columns):
    print("The coefficient for {} is {}".format(col_name, regression_model.coef_[idx]))
```

```python
regression_model.score(x_test, y_test)
```

```python
#  Iteration -2
from sklearn.preprocessing import PolynomialFeatures
from sklearn import linear_model

poly = PolynomialFeatures(degree=2, interaction_only=True)
x_train_ = poly.fit_transform(x_train)

x_test_ = poly.fit_transform(x_test)

poly_clf = linear_model.LinearRegression()

poly_clf.fit(x_train_, y_train)

y_pred = poly_clf.predict(x_test_)

#print(y_pred)

print(poly_clf.score(x_test_, y_test))
```
